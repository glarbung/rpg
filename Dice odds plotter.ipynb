```python
import numpy as np
import matplotlib.pyplot as plt

def successes(dice, tn):
    
    success = 0
    n = 1
    while n < dice +1:
        x = np.random.randint(1,dicetype+1)
        if x > TN-1:
            success = success +1
        
        if explosion == 1:
            if x == dicetype:
                recursion = successes(1,tn)
                if recursion > 0:
                    success = success + recursion
        n += 1
    
    return success

def throw(TN, num_dice, repeats):
    p = np.zeros((repeats+1,1),dtype=int)

    i = 0
    while i < repeats +1:
        p[i] = successes(num_dice,TN)
        i += 1
    fig = plt.hist(p,bins=np.max(p),density=1)
    plt.title(r'%iD and TN%i' %(num_dice,TN))
    plt.show()
```

```python
global explosion
global dicetype

min_haitari = 1
max_haitari = 6
TN = 8
repeats = 100000
explosion = 1
dicetype = 10

i = min_haitari
while i < max_haitari+1:
    throw(TN,i,repeats)
    i += 1    
```
